# Kết nối SQL với Excel phục vụ báo cáo tự động

## From SQL server

- Mở `Excel >> Data >> From other sources >> From SQL server`. Gõ tên địa chỉ Server names, user, password

![](Images/connectsql1.png)
  
- Chọn `Next` liên tục đến khi hiển thị ra thông tin như ảnh dưới đây:
  
  ![](Images/connectsql2.png)
  
- Chọn `Finish` để ra màn hình chọn `Properties`

![](Images/connectsql3.png)
  
- Chọn `Definition` >> phần `Command type` chọn SQL >>  gõ câu lệnh SQL truy vấn dữ liệu

![](Images/connectsql4.png)
  
> **Lưu ý**: Có thể chọn `Save password` để khi gửi file cho client có thể tự refresh file khi dữ liệu trên SQL thay đổi.

## Kết nối SQL với Microft Query

- `From other sources >> From Microsoft Query >> New Data source`

![](Images/connectsql5.png)
  
- Chọn tên đặt cho source kết nối, **chọn driver là SQL server** để đảm bảo client có thể refresh được dữ liệu.

![](Images/connectsql6.png)
  
- Chọn `Connect` >> gõ tên Server muốn kết nối >> Bỏ chọn `Use Trusted Connection` để chọn User và Password. Chọn `save Password` cho những lần kết nối sau

![](Images/connectsql7.png)
  
- Chọn `All column` trong phần `Query Wizard`

![](Images/connectsql8.png)
  
- Chọn `Next` liên tục đến khi hiển thị ra đến phần chọn `Return data to Microsoft >>  Finish`.

![](Images/connectsql9.png)
  
- Phần `Properties >> Command` text gõ câu lệnh SQL truy vấn, **phần parameter cần để =? **
    
![](Images/connectsql10.png)

- Điền `parameter value`, lựa chọn vị trí ô để gõ dữ liệu `filter` >> chọn ô `User this value` và `Refresh automatically`.
  
![](Images/connectsql11.png)

## Chỉnh sửa cậu lệnh truy vấn SQL sau khi đã tạo connection

- Chọn: `Data >> Connections >> Properties >> Definition` để sửa câu lệnh truy vấn ở phần `Command text`.

> **Lưu ý**: Sử dụng kết nối SQL và Excel sẽ cho phép tạo các báo cáo động trong Excel khi làm việc với SQL
